## Introduction

The ability to recognize and classify handwritten numbers is a fundamental problem in the field of machine learning. This project delves into the world of digit classification by implementing a python-based neural network algorithm to solve simple digit classifications. This project implements a multi-layer perceptron (MLP) with basic backpropogation (BP) to classify handwritten digits from 0 - 9 from a 28 x 28 pixelated image. A 2 layer mlp with 784 input neurons, 50 neurons in the hidden layer, and 1 output neuron was implemented to classify single digits. In this document, this architecture only classified the handwritten "1"s and "3"s to confirm that the model works with high accuracy. To classify all the digits, a 2 layer mlp with 784 input neurons, 50 neurons in the hidden layer, and 10 output neurons was used. The datasets have been modified so that the 28 x 28 matrix is now a 784 x 1 column vector. More information on how the datasets were modified along with project instructions and important documents are in the appendix.

#### Imports and Data Preparation

```python
# Imports 
import sys
sys.path.append('/Users/Andy/Documents/git/digitClassifier/mlpToolbox')

import pandas as pd
import numpy as np
import predbp1p
import mlpBP

# Data 
TRNXX = pd.read_csv(r'dataset/TRN_MNIST/TRNXX.csv', header = None)
TRNY1 = pd.read_csv(r'dataset/TRN_MNIST/TRNY1.csv', header = None)
TRNY3 = pd.read_csv(r'dataset/TRN_MNIST/TRNY3.csv', header = None)
TRN10Y = pd.read_csv(r'dataset/TRN_MNIST/TRN10Y.csv', header = None)

RSVXX = pd.read_csv(r'dataset/RSV_MNIST/RSVXX.csv', header = None)
RSVY1 = pd.read_csv(r'dataset/RSV_MNIST/RSVY1.csv', header = None)
RSVY3 = pd.read_csv(r'dataset/RSV_MNIST/RSVY3.csv', header = None)
RSV10Y = pd.read_csv(r'dataset/RSV_MNIST/RSV10Y.csv', header = None)


col_num = 5000

VALXX = RSVXX.iloc[:, :col_num]
TSTXX = RSVXX.iloc[:, col_num:]

VALY3 = RSVY3.iloc[:, :col_num]
TSTY3 = RSVY3.iloc[:, col_num:]

TSTXX.columns = range(col_num)
TSTY3.columns = range(col_num)

VAL10Y = RSV10Y.iloc[:, :col_num]
TST10Y = RSV10Y.iloc[:, col_num:]

TST10Y.columns = range(col_num)

TRNXX = TRNXX.to_numpy()
TRNY1 = TRNY1.to_numpy()
TRNY3 = TRNY3.to_numpy()
TRN10Y = TRN10Y.to_numpy()

RSVXX = RSVXX.to_numpy()
RSVY1 = RSVY1.to_numpy()
RSVY3 = RSVY3.to_numpy()
RSV10Y = RSV10Y.to_numpy()

TSTXX = TSTXX.to_numpy()
TSTY3 = TSTY3.to_numpy()
TST10Y = TST10Y.to_numpy()

VALXX = VALXX.to_numpy()
VALY3 = VALY3.to_numpy()
VAL10Y = VAL10Y.to_numpy()
```

## 1. "1" Detector

### 1.1 "1" Detector: Training

```python
alpha1 = 0.03
alpha2 = 0.005
maxep = 60
hidden_layer_PEs = 50
```

#### 1.1.1 "1" Detector: 1st Training Run

```python
W1_final_alpha1, W2_final_alpha1, b1_final_alpha1, b2_final_alpha1, AE2_final_alpha1 = mlpBP.mlpBP(TRNXX, TRNY1, hidden_layer_PEs, alpha1, maxep)
print("Final MSE", AE2_final_alpha1)
```

#### 1.1.2 "1" Detector: 2nd Training Run

```python
W1_final_alp2, W2_final_alp2, b1_final_alp2, b2_final_alp2, AE2_final_alp2 = mlpBP.mlpBP(TRNXX, TRNY1, hidden_layer_PEs, alpha2, maxep)
print("Final MSE", AE2_final_alp2)
```

#### 1.1.3 "1" Detector: 3rd Training Run - Stopping at less epochs with the best alpha

```python
less_epochs1 = 15
W1_final, W2_final, b1_final, b2_final, AE2_final = mlpBP.mlpBP(TRNXX, TRNY1, hidden_layer_PEs, alpha2, less_epochs1)
print("Final MSE", AE2_final)
```

### 1.2 "1" Detector: Predicting

```python
cols_p = np.shape(RSVXX)[1]

hits = 0
for i in range(cols_p):
    avg2, hit = predbp1p.predbp1p(W1_final, W2_final, b1_final, b2_final, RSVXX[:,i], RSVY1[:,i])
    hits += hit
print("Avg2:", avg2)
print("Correct Classifications:", hits)
print("Incorrect Classification", cols_p - hits)
print("Accuracy:", hits / cols_p)
```

## 2. "3" Detector

### 2.1 "3" Detector: Training and Validation

```python
val_alpha1 = 0.01
val_alpha2 = 0.001
```

#### 2.1.1 "3" Detector: 1st Training Run

```python
W1_val_dig3_alp1, W2_val_dig3_alp1, b1_val_dig3_alp1, b2_val_dig3_alp1, AE2_val_dig3_alp1 = mlpBP.mlpBP(TRNXX, TRNY3, hidden_layer_PEs, val_alpha1, maxep, VALXX=VALXX, VALY=VALY3)
print("Final MSE", AE2_val_dig3_alp1)
```

#### 2.1.2 "3" Detector: 2nd Training Run

```python
W1_val_dig3_alp2, W2_val_dig3_alp2, b1_val_dig3_alp2, b2_val_dig3_alp2, AE2_val_dig3_alp2 = mlpBP.mlpBP(TRNXX, TRNY3, hidden_layer_PEs, val_alpha2, maxep, VALXX=VALXX, VALY=VALY3)
print("Final MSE", AE2_val_dig3_alp2)
```

#### 2.1.3 "3" Detector: 3rd Training Run - Stopping at less epochs with best alpha

```python
less_epochs2 = 20
W1_val_dig3, W2_val_dig3, b1_val_dig3, b2_val_dig3, AE2_val_dig3 = mlpBP.mlpBP(TRNXX, TRNY3, hidden_layer_PEs, val_alpha2, less_epochs2, VALXX=VALXX, VALY=VALY3)
print("Final MSE", AE2_val_dig3)
```

### 2.2 "3" Detector: Predicting

```python
cols_p = np.shape(TSTXX)[1]

hits = 0
for i in range(cols_p):
    
    avg2, hit = predbp1p.predbp1p(W1_val_dig3, W2_val_dig3, b1_val_dig3, b2_val_dig3, TSTXX[:,i], TSTY3[:,i])
    hits += hit
print("Avg2:", avg2)
print("Correct Classifications:", hits)
print("Incorrect Classification", cols_p - hits)
print("Accuracy:", hits / cols_p)
```

## 4. Observation and Conclusions

In this document, the application of artificial neural networks, specifically the Multi-Layer Perceptron with basic backpropagation was implemented with the purpose of classifying handwritten digits. The primary objective was to classify a handwritten "1" and "3", and then create a more complex network to classify all handwritten digits from 0 to 9. To classify the handwritten "1"s and "3"s, a 2 layer mlp with 784 input neurons, 50 neurons in the hidden layer, and 1 output neuron was created. To classify all the digits from 0 to 9, the network was altered so that it has 10 output neurons. These architectures achieved 99% accuracy when classifying the handwritten "1", 98% when classifying the handwritten 3, and 90% accuracy when classifying all digits. 

All networks were initially trained with 60 epochs of training to test different alpha values and to analyze and determine at what number of epochs would achieve the optimal balance between computation time and accuracy.

When classifying the handwritten "1", the architecture achieved around 99% accuracy. After 3 training runs, it was concluded that for classifying the number "1", a suitable alpha is 0.005 and only 15 epochs of training to achieves high accuracy and relatively low computation time.

When classifying the handwritten "3", the architecture achieved around 98% accuracy. After 3 training runs, it was concluded that for classifying the number "3", a suitable alpha is 0.001 and only 20 epochs of training to achieves high accuracy and relatively low computation time.

In the case of the network with 10 output neurons to classy all the digits, the architecture achieved around 90% accuracy. After 3 training runs, it was concluded that a suitable alpha is 0.001 and only 30 epochs of training to achieves high accuracy and relatively low computation time.
